```python
stop_words = set(stopwords.words("english"))
filter_list=[]
for i in w:
  word_tokens = word_tokenize(i)
  filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
  filtered_sentence = []
  for w in word_tokens:
    if w not in stop_words:
      filtered_sentence.append(w)
  filter_list.append(filtered_sentence)
print(filter_list)
```

```python
stemmer = PorterStemmer()
for i in filter_list:
  for word in i:
    print(stemmer.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
l=[]
for i in filter_list:
    temp = []
    for word in i:
        temp.append(lemmatizer.lemmatize(word))
    l.append(temp)
print(l)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# create the vocabulary
vectorizer = CountVectorizer()
filter_list = [' '.join(words) for words in filter_list]
# fit the vocabulary to the text data
vectorizer.fit(filter_list)

# create the bag-of-words model
bow_model = vectorizer.transform(filter_list)

# print the bag-of-words model
print(bow_model)
```

```python
import nltk

# create the vocabulary
vocab = set()

# create the bag-of-words model
bow_model = []

for text in filter_list:
    # create a dictionary to store the word counts
    word_counts = {}

    # tokenize the text
    tokens = nltk.word_tokenize(text)

    # update the vocabulary
    vocab.update(tokens)

    # count the occurrences of each word
    for word in tokens:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1

    # add the word counts to the bag-of-words model
    bow_model.append(word_counts)
print(bow_model)
```
