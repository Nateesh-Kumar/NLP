```python
import whisper
from jiwer import wer

# Load the Whisper model
model = whisper.load_model("base")

# Transcribe the audio file
transcription = model.transcribe("/content/drive/MyDrive/Colab Notebooks/DATASET/dataset1/audio dataset/Recording.mp3")

# The transcribed text by Whisper
predicted_text = transcription["text"]

# Reference transcript (this should be the correct/expected transcription)
# Example: Replace this with the actual correct transcription text
reference_text = "Hello everyone, this audio is for demo purpose to show how best per transforms the audio data into the text. Thank you."

# Calculate Word Error Rate (WER)
error_rate = wer(reference_text, predicted_text)

# Display the transcribed text and WER
print("Transcribed Text:", predicted_text)
print(f"Word Error Rate (WER): {error_rate * 100:.2f}%")
```
